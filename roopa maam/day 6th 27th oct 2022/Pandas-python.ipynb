```python
#DataFrame Aggregate Functions
info1 = pd.DataFrame({"x":[25,15,12,19],   
                    "y":[47, 24, 17, 29]})     
# Create second Dataframe using dictionary   
info2 = pd.DataFrame({"x":[25, 15, 12],   
                    "y":[47, 24, 17],    
                    "z":[38, 12, 45]})   
# append info2 at end in info1   
print(info1.agg(['sum','min','max']) ) 
print(info1)
```

```python
#Assign function
# Create an empty dataframe  
info = pd.DataFrame()  
  
# Create a column  
info['ID'] = [101, 102, 103]
print(info)
info1=info.assign(Name = ['Smith', 'Parker', 'John'])  
print(info1)
info1=info1.assign(Mark1=[79,67,89])
info1=info1.assign(Mark2=[79,67,89])
info1=info1.assign(Total=lambda x: x.Mark1+x.Mark2)
print(info1)
print(info1.count())
```

```python
#Head function
d1=pd.DataFrame({'num1': np.arange(0,100,5,int) }) 
d2=pd.DataFrame({'num2': np.arange(0,100,5,int) }) 
print(d1.head())
print(d1.mean())
```

```python
#merge and rename function
left = pd.DataFrame({  
   'id':[1,2,3,4,5],  
   'Name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'],  
   'subject_id':['sub1','sub2','sub4','sub6','sub5']})  
right = pd.DataFrame({  
    'id':[1,2,3,4,5],  
   'Name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'],  
   'subject_id':['sub2','sub4','sub3','sub6','sub5']})  
d3=pd.merge(left,right,on='id')  
print(d3)
d3.rename(columns = {'Name_x':'Name'}, inplace = True)
print(d3)
```

```python
#sample function
d1=pd.DataFrame({'num1': np.arange(0,100,5,int) }) 
d2=d1.sample(n=3)
print(d2)
```

```python
#shift function
d1=pd.DataFrame({'num1': np.arange(0,100,5,int) }) 
d2=d1.shift(periods=2,axis=1,fill_value=70)
#d2=d1.shift(periods=2)
print(d2)
```

```python
#Sort function
info = pd.DataFrame(index=[3,4,1,2])
info = pd.DataFrame({'col1':[7,1,8,3],'col2':[8,12,4,9]})  
print(info) 
info1=info.sort_values(by='col1')
print(info1)
#transpose function
print(info1.transpose())
#Adding columns
info1['colum3']=24
print(info1)
#converting to numpy array
a=info1.to_numpy()
print(a)
```
