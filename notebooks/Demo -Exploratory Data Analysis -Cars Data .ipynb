## Scatterplot
## Use scatter plots to find the correlation between two numerical variables. 
## Scatter plots are plotted between Horsepower and Price 
## With the plot given below, we can easily draw a trend line. 
## These features provide a good scattering of points.

```python
# Plotting a scatter plot
fig, ax = plt.subplots(figsize=(10,6))
ax.scatter(Carsdata["HP"], Carsdata["Price"])
ax.set_xlabel("HP")
ax.set_ylabel("Price")
plt.show()
```

```python
# Plotting a scatter plot
fig, ax = plt.subplots(figsize=(10,6))
ax.scatter(Carsdata["Make"], Carsdata["Price"])
ax.set_xlabel("Make")
ax.set_ylabel("Price")
plt.show()
```
