```python
def mean_normalize(series):
    norm_series = (series-series.mean())/(series.max()-series.min())
    return norm_series
```

```python
series = pd.Series([5,192,20019,12,209])
mean_normalize(series)
```

```python
# fft
import numpy as np

def fast_forier_transform(star):
  fft_star=np.fft.fft(star, n=len(star))
  return np.abs(fft_star)
```
